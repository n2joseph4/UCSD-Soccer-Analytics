```python
dots =fullHtml.find("teamstats__Dots-module__dots")
dotsData = fullHtml[dots:]

labels = dotsData.split('<div class="teamstats__Dots-module__dot')
positions = []
for label in labels[1:]:
    left = label.find("left:")
    left = int(label[left+6:left+8].replace("%",""))/100*120
    top = label.find("top:")
    top = int(label[top+5:top+7].replace("%",""))/100*80

    positions.append([left,top,"module__successful" in label])
positions = pd.DataFrame(positions,columns=["x1","y1","won"])
```

```python
won = positions[positions.get("won")]
lost = positions[~positions.get("won")]
pitch = Pitch(positional=True, axis=True,label=True)
fig, ax = plt.subplots(figsize = (10,6))
pitch.draw(ax=ax)
kde = sns.kdeplot(x=lost.get("x1"),y=lost.get("y1"),fill = True,shade_lowest = False,alpha = .5,n_levels = 20,cmap = "magma")
for i in range(lost.shape[0]):
    row = lost.iloc[i]
    if row.get("won"):
        color = "og"
    else:
        color = "ro"
    plt.plot(row.get("x1"),row.get("y1"),color)
plt.title("UCSD Offensive Duels Lost")
```
