```python
arrows = fullHtml.find("teamstats__Arrows-module__arrows")
dots =fullHtml.find("teamstats__Dots-module__dots")
arrowsData = fullHtml[arrows:dots]
dotsData = fullHtml[dots:]

labels = dotsData.split('<div class="teamstats__Dots-module__dot')
positions = []
for label in labels[1:]:
    left = label.find("left:")
    left = int(label[left+6:left+8].replace("%",""))/100*120
    top = label.find("top:")
    top = int(label[top+5:top+7].replace("%",""))/100*80
    positions.append([left,top])
positions = pd.DataFrame(positions,columns=["x1","y1"])
arrowLabels = arrowsData.split('<g class=""><defs>')
arrows = []
for label in arrowLabels[1:]:
    x1 = label.find("x1")
    y1 = label.find("y1")
    x2 = label.find("x2")
    y2 = label.find("y2")
    coords = label.split("=")[17:]
    coordinates = []
    for i in range(4):
        if i%2==0:
            divided = 710.2359999999999/120
        else:
            divided = 470/80
        val = coords[i]
        val = float(val.split('"')[1])/divided
        coordinates.append(val)
    arrows.append(coordinates)
arrows = pd.DataFrame(arrows,columns=["x1",'y1',"x2","y2"])
pos = positions.to_numpy()
arr = arrows.get(["x2","y2"]).to_numpy()
merged = np.concatenate((pos,arr))
merged = pd.DataFrame(merged,columns=['x','y'])
```
