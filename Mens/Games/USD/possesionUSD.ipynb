```python
import pandas as pd
import os
```

```python
def removeUnamed(df):
    columnList = df.columns.tolist()
    newColumnList = columnList.copy()
    i = 0
    while i<len(newColumnList):
        if "Unnamed" in newColumnList[i]:
            stats = newColumnList[i-1].split(" / ")
            newColumnList[i-1] = stats[0]
            newColumnList[i] = " ".join(stats)
        i+=1
    columnsRenamed = dict(zip(columnList,newColumnList))
    df = df.rename(columns=columnsRenamed)
    return df
```

```python
files = os.listdir("Data/Players")
```

```python
playersData = pd.read_excel("Data/Players/"+files[0])
playersData = removeUnamed(playersData)
playersData = playersData.assign(Player = files[0].split(".")[0])
for file in files:
    df = pd.read_excel('Data/Players/'+file)
    df = removeUnamed(df)
    df = df.assign(Player = file.split(".")[0])
    playersData = pd.concat([playersData,df])
playersData = playersData[playersData.get("Match").apply(lambda match: "UC San Diego" in match)]
playersData = playersData[playersData.get("Date").apply(lambda date: '2024' in date)]
playersData.drop_duplicates(inplace=True)
usd = playersData[playersData.get("Match").apply(lambda match: "San Diego Toreros" in match)]
usd=usd.set_index("Player")
usd.get(["Passes accurate","Passes","Offensive duels won","Offensive duels"])
playersData
```

```python
usd = usd.assign(Possession = (usd.get("Passes accurate")+usd.get("Offensive duels won"))/(usd.get("Passes")+usd.get("Offensive duels")))
usd = usd.assign(PossessionActions = (usd.get("Passes")+usd.get("Offensive duels")))
usd.sort_values("Possession",ascending=False).get(["Possession", "PossessionActions"])
```

```python
players = playersData.drop(["Match","Competition","Date","Position"],axis = 1).groupby("Player").sum()
players = players.assign(Possession = (players.get("Passes accurate")+players.get("Offensive duels won"))/(players.get("Passes")+players.get("Offensive duels")))
players = players.assign(PossessionActions = (players.get("Passes")+players.get("Offensive duels")))
players.sort_values("Possession",ascending=False).get(["Possession", "PossessionActions"])
```
