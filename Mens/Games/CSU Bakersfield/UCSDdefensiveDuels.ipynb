```python
dots =fullHtml.find("teamstats__Dots-module__dots")
dotsData = fullHtml[dots:]

labels = dotsData.split('<div class="teamstats__Dots-module__dot')
positions = []
for label in labels[1:]:
    left = label.find("left:")
    left = int(label[left+6:left+8].replace("%",""))/100*120
    top = label.find("top:")
    top = int(label[top+5:top+7].replace("%",""))/100*80
    player = int(label.split("span")[-2].replace(">","").replace("<","").replace("/",""))
    positions.append([left,top,"module__successful" in label, player])
positions = pd.DataFrame(positions,columns=["x1","y1","won","player"])
duelsOwnHalf = positions[positions.get("x1")<60]
def assignSide(y1):
    if y1<40:
        return "Left"
    else:
        return "Right"
duelsOwnHalf = duelsOwnHalf.assign(Side = duelsOwnHalf.get("y1").apply(assignSide))
left = duelsOwnHalf[duelsOwnHalf.get("Side")=="Left"]
right = duelsOwnHalf[duelsOwnHalf.get("Side")=="Right"]
won = positions[positions.get("won")]
lost = positions[~positions.get("won")]
```

```python
print(duelsOwnHalf.get("won").mean(),left.get("won").mean(),right.get("won").mean())
```
