```python
dots =fullHtml.find("teamstats__Dots-module__dots")
dotsData = fullHtml[dots:]

labels = dotsData.split('<div class="teamstats__Dots-module__dot')
positions = []
for label in labels[1:]:
    left = label.find("left:")
    left = int(label[left+6:left+8].replace("%",""))/100*120
    top = label.find("top:")
    top = int(label[top+5:top+7].replace("%",""))/100*80
    player = int(label.split("span")[-2].replace(">","").replace("<","").replace("/",""))
    positions.append([left,top,"module__successful" in label, player])
positions = pd.DataFrame(positions,columns=["x1","y1","won","player"])
duelsOwnHalf = positions[positions.get("x1")<60]
def assignSide(y1):
    if y1<40:
        return "Left"
    else:
        return "Right"
duelsOwnHalf = duelsOwnHalf.assign(Side = duelsOwnHalf.get("y1").apply(assignSide))
left = duelsOwnHalf[duelsOwnHalf.get("Side")=="Left"]
right = duelsOwnHalf[duelsOwnHalf.get("Side")=="Right"]
```

```python
print(duelsOwnHalf.get("won").mean(),left.get("won").mean(),right.get("won").mean())
```

```python
pitch = Pitch(positional=True, axis=True,label=True)
fig, ax = plt.subplots(figsize = (10,6))
pitch.draw(ax=ax)
kde = sns.kdeplot(x=positions.get("x1"),y=positions.get("y1"),fill = True,shade_lowest = False,alpha = .5,n_levels = 15,cmap = "magma")
for i in range(positions.shape[0]):
    row = positions.iloc[i]
    if row.get("won"):
        color = "og"
    else:
        color = "ro"
    plt.plot(row.get("x1"),row.get("y1"),color)
plt.title("UCSD Defensive Duels")
plt.show()
won = positions[positions.get("won")]
lost = positions[~positions.get("won")]
pitch = Pitch(positional=True, axis=True,label=True)
fig, ax = plt.subplots(figsize = (10,6))
pitch.draw(ax=ax)
kde = sns.kdeplot(x=lost.get("x1"),y=lost.get("y1"),fill = True,shade_lowest = False,alpha = .5,n_levels = 15,cmap = "magma")
for i in range(lost.shape[0]):
    row = lost.iloc[i]
    if row.get("won"):
        color = "og"
    else:
        color = "ro"
    plt.plot(row.get("x1"),row.get("y1"),color)
plt.title("UCSD Defensive Duels lost")
plt.show()
pitch = Pitch(positional=True, axis=True,label=True)
fig, ax = plt.subplots(figsize = (10,6))
pitch.draw(ax=ax)
kde = sns.kdeplot(x=won.get("x1"),y=won.get("y1"),fill = True,shade_lowest = False,alpha = .5,n_levels = 15,cmap = "magma")
for i in range(won.shape[0]):
    row = won.iloc[i]
    if row.get("won"):
        color = "og"
    else:
        color = "ro"
    plt.plot(row.get("x1"),row.get("y1"),color)
plt.title("UCSD Defensive Duels Won")
plt.show()
```

```python
bins = (6,5)
pitch = Pitch(positional=True)
fig, ax = pitch.draw(figsize=(16, 11), constrained_layout=True, tight_layout=False)
fig.set_facecolor('#22312b')
# plot the heatmap - darker colors = more passes originating from that square
bs_heatmap = pitch.bin_statistic(positions.x1, positions.y1, statistic='count', bins=bins)
successful_bins = pitch.bin_statistic(won.x1, won.y1, statistic='count', bins=bins)
unsuccessful_bins = pitch.bin_statistic(lost.x1, lost.y1, statistic='count', bins=bins)
accuracies = successful_bins["statistic"]/(successful_bins["statistic"]+unsuccessful_bins["statistic"])
bs_heatmap["statistic"]=accuracies
hm = pitch.heatmap(bs_heatmap, ax=ax, cmap='Greens',alpha = .6)

labels2 = pitch.label_heatmap(bs_heatmap, color='black', fontsize=18,
                              ax=ax, ha='center', va='center',
                              str_format='{:.0%}',)
plt.title("Defensive Duels Accuracy on Pitch",color = "white",fontsize = 18)
```
